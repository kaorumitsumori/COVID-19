## Read json file from Tokyo COVID-19 Task Force website

```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt

tokyo_data = pd.read_json('/Users/kaorumitsumori/Desktop/covid19/data/data.json')
tokyo_data["patients"]["data"]
```

## Convert into DataFrame and quick look

```python
df_tokyo_data = pd.DataFrame(tokyo_data["patients"]["data"])
df_tokyo_data.head(10)
```

```python
df_tokyo_data.tail(10)
```

## Total patients' number in Tokyo

```python
patients_number = len(df_tokyo_data)

print('The number of patients of COVID-19 in Tokyo by 15 March 2020 is',patients_number)
```

## Pie chart of patients (ages)

```python
import japanize_matplotlib

x = df_tokyo_data['年代'].value_counts(ascending=True)
label_index = df_tokyo_data['年代'].value_counts(ascending=True).index
plt.figure(figsize=(10,10))
plt.pie(x, labels= label_index, startangle=90, labeldistance=1,rotatelabels=False,autopct='%.2f%%')
plt.show()
```
