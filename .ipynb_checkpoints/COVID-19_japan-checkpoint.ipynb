# Report on COVID-19 entirely in Japan

## Read json file and quick look

```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt

japan_data = pd.read_csv('/Users/kaorumitsumori/Desktop/COVID-19/data_japan20200315.csv')
japan_data.info()
```

```python
japan_data.head(10)
```

## Total number by symptom

```python
total = japan_data["累計"].iloc[-1]
d = japan_data["死者合計"].dropna().iloc[-1]
di = japan_data["退院数累計"].dropna().iloc[-1]

print(japan_data["確定日"].iloc[-1],"Current data below;")
print('Total number of infected people around nation:',total)
print('Number of dead people:',d)
print('Number of people who left hospital:',di)
print('Mortality：{}%'.format((d/total)*100))
print('Ratio of leaving hospital：{}%'.format((di/total)*100))
```

## Increase in infected people

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(japan_data['確定日'],japan_data['累計'])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, fontsize=10);
```

## Increase in dead people

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.bar(japan_data['確定日'],japan_data["死者合計"])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, fontsize=10);
```

# Comparison with infected and dead

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.bar(japan_data['確定日'],japan_data["死者合計"])
ax.plot(japan_data['確定日'],japan_data['累計'])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, fontsize=10);
```

## Pie chart of patients (by age)

```python
x = japan_data['年代'].value_counts(ascending=True)
label_index = japan_data['年代'].value_counts(ascending=True).index
plt.figure(figsize=(10,10))
plt.pie(x, labels= label_index, startangle=90, labeldistance=1,rotatelabels=False,autopct='%.2f%%')
plt.legend()
plt.show()
```

## Pie chart of patients (by gender)

```python
x = japan_data['性別'].value_counts(ascending=True)
label_index = japan_data['性別'].value_counts(ascending=True).index
plt.figure(figsize=(10,10))
plt.pie(x, labels= label_index, startangle=90, labeldistance=1,rotatelabels=False,autopct='%.2f%%')
plt.legend()
plt.show()
```

## Numbers of infected by prefectures

```python
grouped = japan_data['人数'].groupby(japan_data['居住都道府県'])
total_pref = grouped.sum()
total_pref = total_pref.sort_values(ascending=False) 
total_pref = pd.DataFrame(total_pref)
total_pref
```
